In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
# Loading from url not working. Loading the file locally. 
#df_mkt = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv")
df_mkt = pd.read_csv("mkt_analysis_dataset.csv")
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
if 'unnamed:_0' in df_mkt.columns:
    df_mkt = df_mkt.drop(columns='unnamed:_0')
df_mkt.head()
```

```python
df_mkt.columns
```

```python
display((df_mkt.duplicated().sum())/df_mkt.shape[0]*100)
display(df_mkt.shape)
```

```python
"""
Creating a copy of the df, dropping rows from the dataset if customer and CLV 
are duplicated and printing the % of the original dataframe which is dropped 
as a result of the duplicated values in customer and clv. 
"""
test_df = df_mkt.copy()
test_df = test_df.drop_duplicates(subset=
                                  ['customer', 'customer_lifetime_value']
                                 , keep='first')
new_shape = test_df.shape[0]
old_shape = df_mkt.shape[0]
print('Number of removed duplicated rows = ' + str(old_shape - new_shape))
print('This is ' + str(round(((old_shape - new_shape)/old_shape)*100, 2)) 
                       + '% of the previous dataset')
```

```python
""" 
Pivoting the table to show the volume of customer_lifetime_value
based on sales channel, and rounding the decimal points to 2.
"""

test_df_pivoted = round(test_df.pivot_table(
              columns='sales_channel', values='customer_lifetime_value', 
    aggfunc='sum'), 2)
test_df_pivoted
```

```python
"""
Using a groupby function to see if the values are the same if achieved 
by a different method. 
Upon inspection, values are the same. 
"""
test_df_grouped = test_df.groupby(['sales_channel'])['customer_lifetime_value'].sum().reset_index()
test_df_grouped['customer_lifetime_value'] = round(test_df_grouped['customer_lifetime_value'], 2)
test_df_grouped
```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
test_df_pivot_2 = test_df.pivot_table(columns=['education', 'gender'], 
                    values='customer_lifetime_value', aggfunc='mean')
test_df_pivot_2
```

```python
"""
Education values are ordinal. Changing order to (ascending): 
- Highschool or Below
- College
- Bachelor
- Master
- Doctor
"""

test_df_pivot_2.columns
```

```python
cols = [('High School or Below', 'F'),('High School or Below', 'M'),
       ('College', 'F'), ('College', 'M'),
       ('Bachelor', 'F'),('Bachelor', 'M'),
        ('Master', 'F'),('Master', 'M'),
       ('Doctor', 'F'),('Doctor', 'M')]
```

```python
# Reordering columns in dataframe. 
test_df_pivot_2 = test_df_pivot_2[cols]
test_df_pivot_2
```

### Insights:
- Customers with either college or doctorate level educations tend to generate lower revenue for the insurance company, both men and women. 
- Females with a high school education tend to generate the highest amount of revenue out of all demographics. 
- Men with doctorate level education tend to generate the lowest revenue. 
- The highest revenue generating demographic are customers with masters education, shortly followed by high school educated customers. 
- The largest gap within the demographics themselves is between male and female highschool students with a difference of ~568.
